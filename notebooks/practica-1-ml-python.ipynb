(b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```python
#Create a 2-dimensional array with random integer values
array = np.random.randint(low=1, high=10, size=(3, 4))
print(array)

#Calculate the sum of each row
filas = np.sum(array, axis=1)

#Calculate the sum of each column
columnas = np.sum(array, axis=0)

print("Suma de filas:")
print(filas)

print("Suma de columnas:")
print(columnas)
```

# 3. Data Import


(a) Download the Echocardiogram data set and import it to a data frame. Read the information on the
data set and find out how missing values are represented and make sure that they are properly
represented.

```python
dataframe = pd.read_csv('/kaggle/input/echocardiogram3/echocardiogram.data', delimiter=',')
dataframe

```

```python
dataframe.info()
```

```python
#Los valores faltantas se represetan con una ?
#Para representarlos correctamente seran reemplazados con NaN
#Reemplazar los valores representados como "?" con NaN
dataframe = dataframe.replace("?", np.nan)

# Mostrar el DataFrame resultante
print(dataframe)
```

(b) Assign the attributes with meaningful names. You can look for this information on the same
webpage.

```python
#Los nombres significativos se encuentran en el archivo echocardiogram.names
attribute_names = [
    'survival',
    'still-alive',
    'age-at-heart-attack',
    'pericardial-effusion',
    'fractional-shortening',
    'epss',
    'lvdd',
    'wall-motion-score',
    'wall-motion-index',
    'mult',
    'name',
    'group',
    'alive-at-1'
]
```

```python
#Reemplazar los valores de cada columna con sus respectivos nombres significativos segun la informacion en el archivo .name
dataframe.columns = attribute_names
dataframe
```

(c) According to that same information, is there any redundant or irrelevant attribute that you can
remove? Remove them.

```python
#Asiendo un analisis a los datos, se identifica que existen 3 columnas que no son relevantes o aportan alguna informacion imporante
#Se eliminan las columnas de "name", "group", "mult" y "alive-at-1"
dataframe = dataframe.drop('name', axis=1)
dataframe = dataframe.drop('group', axis=1)
dataframe = dataframe.drop('mult', axis=1)
dataframe = dataframe.drop('alive-at-1', axis=1)
dataframe
```

(d) Is there any data type change that you find useful? Perform it.

```python
#Cambiar las columnas "still-alive" y "pericardial-effusion" a booleano para reducir los recursos de procesamiento de datos
dataframe['still-alive'] = dataframe['still-alive'].astype(bool)
dataframe['pericardial-effusion'] = dataframe['pericardial-effusion'].astype(bool)
```
