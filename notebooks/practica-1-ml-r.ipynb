# 4. Data Manipulation


Load the airquality data set regarding a set of New York Air Quality Measurements.
 

```r
air_quality <- read_csv("/kaggle/input/new-york-air-quality/airquality.csv", show_col_types = FALSE)
air_quality 
```

(n) For which attributes are there missing values?


```r
#Identificar los atributos con valores faltantes
faltantes <- colnames(air_quality)[colSums(is.na(air_quality)) > 0]

#Mostrar los atributos con valores faltantes
print(faltantes)
```

(o) Do all the attributes are in the most suitable data type? Make the changes you find necessary.


```r
#Mostrar informacion sobre el dataset
str(air_quality)
```

```r
#Reemplazar los valores NaN con la media de la columna
#Media de cada columna
media_Ozone <- mean(x=air_quality$Ozone, na.rm=TRUE)
media_Solar.R <- mean(x=air_quality$Solar.R, na.rm=TRUE)

#Reemplazar NaN
air_quality$Ozone[is.na(air_quality$Ozone)] <- media_Ozone
air_quality$Solar.R[is.na(air_quality$Solar.R)] <- media_Solar.R

#Cambiar todos los datos menos los de "wind" a int
air_quality$Ozone <- as.integer(air_quality$Ozone)
air_quality$Solar.R <- as.integer(air_quality$Solar.R)
air_quality$Temp <- as.integer(air_quality$Temp)
air_quality$Month <- as.integer(air_quality$Month)
air_quality$Day <- as.integer(air_quality$Day)
air_quality
```

(p) What period of the year do these records refer to?

```r
# Crear un vector de los nombres de los meses
meses <- c("Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
           "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre")

# Obtener los períodos del año
periodos <- sapply(air_quality$Month, function(mes) meses[mes])

# Mostrar los períodos del año
unique(periodos)
```
